## Generator / Iterator

```python
def my_generator(max=0):
    n = 0
    while n < max:
        yield 2 ** n
        n += 1
    
```

```python
generator = my_generator(10)
n = next(generator)
print(n)
n = next(generator)
print(n)
n = next(generator)
print(n)
n = next(generator)
print(n)
```
